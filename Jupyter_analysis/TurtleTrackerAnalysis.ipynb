# Turtle Tracker Real-Time Analysis

```python
import requests, json

# Server Url
url = "https://turtletracker.cas.mcmaster.ca"
```

# Login to view full data

Request token by email

```python
# Enter your registered email address
email = "bit86@live.com"

# -----------------------------------------------------------------
session = requests.Session()
response = session.get(url+'/login?user=' + email)
cookies = session.cookies.get_dict()
print(json.loads(response.text)["msg"])
```

# Get Data

Get turtle tracker json data

```python
session = requests.Session()
response = session.get(url+'/getData', cookies=cookies)
data = json.loads(response.text)
print(data)
```

Print data table

```python
import pandas as pd
pd.DataFrame(data)
```

```python
from ipyleaflet import Map, basemaps, basemap_to_tiles, Marker, Icon
m = Map(
    basemap=basemap_to_tiles(basemaps.OpenStreetMap.Mapnik),
    center=(43.053619, -80.102545),
    zoom=12
    )

icon = Icon(icon_url='https://turtletracker.cas.mcmaster.ca/turtle-marker.png',
            icon_size=[25, 25], icon_anchor=[0,0])

for d in data:
  marker = Marker(location=[d["lat"]/10e6, d["lon"]/10e6],
                  draggable=False, icon=icon)
  m.add_layer(marker);

m
```
